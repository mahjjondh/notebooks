```python
#!/usr/bin/env python
# coding: utf-8

from selenium import webdriver
import os
import time
from selenium.webdriver.common.keys import Keys
import numpy as np
import pandas as pd
import time
from selenium.webdriver.support.select import Select
import re
```

```python
chromedriver = "chromedriver.exe"
os.environ["webddriver.chromedriver"] = chromedriver
driver = webdriver.Chrome(chromedriver)
```

```python
driver.get("https://www.google.com.br/maps/@-15.8429902,-48.1101308,15z")
```

```python
df = pd.read_excel('parques_brasilia.xlsx','Planilha1')
```

```python
len_parques = len(df['Nome'])
len_parques
```

```python
csv_parques_coord = open('parques_coord.csv','w+')
csv_parques_coord.writelines('nome|url|latitude|longitude|classificacao\n')
```

```python
for item in range(1, len_parques):
    text_search = driver.find_element_by_tag_name('input')
    text_search.clear()
    text_search.click()
    text_search.send_keys(str(df['Nome'][item])+' Distrito Federal')
    text_search.send_keys(Keys.ENTER)
    time.sleep(5)
    url = driver.current_url
    parque_nome = df['Nome'][item]
    print(str(parque_nome)+'\n')
    print(str(url)+'\n')
    lat_long = re.search(r'@(.*?),\d\dz',url).group(1) #add after
    lat_long_split = lat_long.split(',')
    latitude = lat_long_split[0]
    longitude = lat_long_split[1]
    try:
        classificacao = driver.find_elements_by_class_name('section-rating-term')
        classifica = classificacao[1].text
    except:
        classificacao = 'parque'
    
    
    """
    try:
        lat_long = re.search(r'@(.*?),17z',url).group(1) #add after
        lat_long_split = lat_long.split(',')
        latitude = lat_long_split[0]
        longitude = lat_long_split[1]
    except:
        lat_long = re.search(r'@(.*?),14z',url).group(1) #add after
        lat_long_split = lat_long.split(',')
        latitude = lat_long_split[0]
        longitude = lat_long_split[1]
    """    
    csv_parques_coord.writelines(str(parque_nome)+'|'+str(url)+'|'+str(latitude)+'|'+str(longitude)+'|'+str(classifica)+'\n')
    print('latitude: '+str(latitude))
    print('longitude: '+str(longitude)+'\n')
    print('classificação: '+str(classifica)+'\n')
    time.sleep(9)
```

```python
csv_parques_coord.flush()
csv_parques_coord.close()
```

```python
df_parques = pd.read_csv('parques_coord.csv',delimiter='|', encoding='iso-8859-1')
df_parques = df_parques.iloc[0:]
df_parques
```

```python
import folium
```

```python
df_parques['classificacao'].unique()
```

```python
len_p = len(df_parques['nome'])
len_p
```

```python
mapa_parkis = folium.Map(location=[-15.721,-48.077], zoom_start=10)
```

```python
for parque in range(0, len_p):
    if df_parques['classificacao'][parque] == 'Parque':
        folium.Marker(location=[df_parques['latitude'][parque],df_parques['longitude'][parque]], icon=folium.Icon(color='green'), popup=str(df_parques['nome'][parque])+'<br>'+' - '+str(df_parques['classificacao'][parque])).add_to(mapa_parkis)
    elif df_parques['classificacao'][parque] == 'Parque ecológico':
        folium.Marker(location=[df_parques['latitude'][parque],df_parques['longitude'][parque]], icon=folium.Icon(color='darkgreen'), popup=str(df_parques['nome'][parque])+'<br>'+' - '+str(df_parques['classificacao'][parque])).add_to(mapa_parkis)
    elif df_parques['classificacao'][parque] == 'Reserva natural':
        folium.Marker(location=[df_parques['latitude'][parque],df_parques['longitude'][parque]], icon=folium.Icon(color='pink'), popup=str(df_parques['nome'][parque])+'<br>'+' - '+str(df_parques['classificacao'][parque])).add_to(mapa_parkis)
    elif df_parques['classificacao'][parque] == 'Belezas naturais':
        folium.Marker(location=[df_parques['latitude'][parque],df_parques['longitude'][parque]], icon=folium.Icon(color='cadetblue'), popup=str(df_parques['nome'][parque])+'<br>'+' - '+str(df_parques['classificacao'][parque])).add_to(mapa_parkis)
    elif df_parques['classificacao'][parque] == 'Lago':
        folium.Marker(location=[df_parques['latitude'][parque],df_parques['longitude'][parque]], icon=folium.Icon(color='lightblue'), popup=str(df_parques['nome'][parque])+'<br>'+' - '+str(df_parques['classificacao'][parque])).add_to(mapa_parkis)
    elif df_parques['classificacao'][parque] == 'Parque estadual':
        folium.Marker(location=[df_parques['latitude'][parque],df_parques['longitude'][parque]], icon=folium.Icon(color='lightgreen'), popup=str(df_parques['nome'][parque])+'<br>'+' - '+str(df_parques['classificacao'][parque])).add_to(mapa_parkis)
    elif df_parques['classificacao'][parque] == 'Permanentemente fechado':
        folium.Marker(location=[df_parques['latitude'][parque],df_parques['longitude'][parque]], icon=folium.Icon(color='red'), popup=str(df_parques['nome'][parque])+'<br>'+' - '+str(df_parques['classificacao'][parque])).add_to(mapa_parkis)
    else:
        folium.Marker(location=[df_parques['latitude'][parque],df_parques['longitude'][parque]], icon=folium.Icon(color='gray'), popup=str(df_parques['nome'][parque])+'<br>'+' - '+str(df_parques['classificacao'][parque])).add_to(mapa_parkis)
        
```

```python
def export_map(map_file, map_image, map_obj):
    browser = webdriver.Chrome('chromedriver.exe')
    tmpURL = 'file://{path}/{mapfile}'.format(path=os.getcwd(), mapfile=map_file)
    map_obj.save(map_file)
    browser.get(tmpURL)
    time.sleep(5)
    browser.save_screenshot(map_image)
    browser.quit()
```

```python
mapa = 'mapa.html'
imagem = 'imagem.png'
```

```python
export_map(mapa,imagem,mapa_parkis)
```
