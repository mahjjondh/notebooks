```python
from selenium import webdriver
import os
import time
from selenium.webdriver.common.keys import Keys
import numpy as np
import pandas as pd
import time
from selenium.webdriver.support.select import Select
```

```python
import re
import sqlite3
from bs4 import BeautifulSoup
import urllib.request
import requests
```

```python
chromedriver = "chromedriver.exe"
os.environ["webddriver.chromedriver"] = chromedriver
driver = webdriver.Chrome(chromedriver)
```

```python
driver.get("https://www.pciconcursos.com.br/concursos")
```

```python
chaves = ['coloque as palavras chaves aqui']
```

```python
patterns = re.compile('|'.join(r'\b{}\b'.format(s) for s in chaves), re.I)
```

```python
cc = driver.find_elements_by_class_name('cc')
ca = driver.find_elements_by_class_name('ca')
len_cc = len(cc)
```

```python
conn = sqlite3.connect('pci_concursos_db.db') #tb_editais - ID, ORGAO, OBJETO, ABERTURA, PRAZO
cur = conn.cursor()
cur.execute('DELETE FROM tb_pci')
```

```python
urls = ''

for f in range(0,len_cc):
    try:
        cc = driver.find_elements_by_class_name('cc')
        ca = driver.find_elements_by_class_name('ca')
        #print(cc[f].text)
        print('*************************************************************************')
        #print(ca[f].text)
        cabecalho = ca[f].text
        ca[f].click()
        time.sleep(10)
        urls = str(driver.current_url)
        req = urllib.request.Request(urls)
        rs = urllib.request.urlopen(req)
        res = requests.get(urls.format(rs))
        soup = BeautifulSoup(res.content,"html.parser")
        texto = driver.find_elements_by_tag_name('p')
        time.sleep(25)
    except:
        pass
    #textopregex = texto[4:]

    #for t in soup:
        #time.sleep(15)
    print('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
    try:
        #ttotal = '-'.join(t.text).replace('-','')
        #print(str(soup))
        #print(ca[f].text)
        patt = patterns.findall(str(soup))
        print(patterns.findall(str(soup)))
        print(str(urls))
        print(str(cabecalho))
        cur.execute('insert into tb_pci values (?,?,?,?)', (None,str(urls),str(patt).replace("''","").replace('[','').replace(']',''),str(cabecalho)))
        conn.commit()
    except:
        pass
    driver.back()
    print('-------------------------------------------------------------------------')
    time.sleep(25)
```

```python
cur.execute("select * from tb_pci where keywords != ''")
rows = cur.fetchall()
```

```python
arquivo = open('arquivo_editais.html','w+')
arquivo.writelines('<html><body><center>')
bgcolor = '85FFFF'
for row in rows:
    arquivo.writelines('<table border=1px solid, style="width:750px;">')
    print('Concurso: \n\n'+str(row[3])+'\n')
    arquivo.writelines('<tr bgcolor="'+str(bgcolor)+'"><td>'+'Concurso: '+str(row[3])+'</tr></td>')
    print('Vagas para: '+str(row[2])+'\n')
    arquivo.writelines('<tr bgcolor="white"><td>'+'Vagas para: '+str(row[2])+'</tr></td>')
    print('Link: '+str(row[1])+'\n')
    arquivo.writelines('<tr bgcolor="white"><td>'+'<a href='+str(row[1])+'>'+str(row[1])+'</a></tr></td>')
    #bgcolor = bgcolor-10
    arquivo.writelines('</table><br>')
arquivo.writelines('</center></body></html>')
arquivo.flush()
arquivo.close()
print('concursos...')
```
