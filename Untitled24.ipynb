```python
def create_date_table2(start='2000-01-01', end='2050-12-31'):
    df_tempo = pd.DataFrame({"Date": pd.date_range(start, end)})
    df_tempo["Dia"] = df_tempo.Date.dt.weekday_name
    df_tempo["Semana"] = df_tempo.Date.dt.weekofyear
    df_tempo["Quarter"] = df_tempo.Date.dt.quarter
    df_tempo["Ano"] = df_tempo.Date.dt.year
    df_tempo["Semestre"] = (df_tempo.Quarter + 1) // 2
    return df_tempo
```

```python
create_date_table2()
```

```python
df_tempo = create_date_table2()
```

```python
df_tempo.to_sql(name='dim_tempo', con=engine_norgs, if_exists='append', index=False)
```
